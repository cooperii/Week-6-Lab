We can merge our two pandasdataframes using the command .merge()
The general form is dataframe1.merge(dataframe2, on='headername', how='left'

```python
# Run this code to merge the ecology dataframe into the morphology dataframe
MorphoEco = Morpho.merge(Eco, on='species', how='left')
```

```python
# now reverse the code and merge the morphology dataframe into the ecology dataframe
# (call the new combo dataframe "EcoMorpho")
EcoMorpho = Eco.merge(Morpho, on='species', how='left')
```

What is the difference in size between the two merged dataframes?
